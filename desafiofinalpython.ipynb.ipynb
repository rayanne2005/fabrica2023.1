**Questão 2A- Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department)**

```python
dfdepart = df.groupby('Department')[['EmployeeCount']].count()
dfdepart
```

```python
dfdepart.plot(kind='bar', color='green')
```

**Qestão 2B- Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education)**
